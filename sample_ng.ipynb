## Subhalo mass function
### with the modification `R * Ffunc_Yang`

```python
m,dNdlnm = obs.mass_function()
m_MVJ,dNdlnm_MVJ = obs_MVJ.mass_function()
m_LMSV,dNdlnm_LMSV = obs_LMSV.mass_function()
m_LMSVQ,dNdlnm_LMSVQ = obs_LMSVQ.mass_function()
m_DMNP,dNdlnm_DMNP = obs_DMNP.mass_function()
m_DMNPF,dNdlnm_DMNPF = obs_DMNPF.mass_function()
plt.figure(figsize=(6,5))
plt.loglog(m,m*dNdlnm)
plt.loglog(m_MVJ,m_MVJ*dNdlnm_MVJ)
plt.loglog(m_LMSV,m_LMSV*dNdlnm_LMSV)
plt.loglog(m_LMSVQ,m_LMSVQ*dNdlnm_LMSVQ)
plt.loglog(m_DMNP,m_DMNP*dNdlnm_DMNP)
plt.loglog(m_DMNPF,m_DMNPF*dNdlnm_DMNPF)
plt.xlim(1.e-7,3.e16)
plt.ylim(1.e7,1.e16)
plt.xlabel(r'$m$ [$M_{\odot}$]')
plt.ylabel(r'$m^{2}dN_{\rm sh}/dm$ [$M_{\odot}$]')
plt.legend(['Gaussian','MVJ, $f_{NL}=2000$','LMSV, $f_{NL}=2000$','LMSVQ, $f_{NL}=2000$','DMNP, $f_{NL}=2000$','DMNPF, $f_{NL}=2000$'])
plt.title(r'$M_{200}=3 \times 10^{15}M_{\odot}$')
plt.show()
```
