```python
import sashimi_c
import sashimi_c_ng
import matplotlib.pyplot as plt
```

```python
obs = sashimi_c.subhalo_observables(M0_per_Msun=1.e14)
obs_MVJ = sashimi_c_ng.subhalo_observables(M0_per_Msun=1.e14, F_NL=5000, NG_model="MVJ")
obs_LMSV = sashimi_c_ng.subhalo_observables(M0_per_Msun=1.e14, F_NL=5000, NG_model="LMSV")
```

## Subhalo mass function

```python
m,dNdlnm = obs.mass_function()
m_MVJ,dNdlnm_MVJ = obs_MVJ.mass_function()
m_LMSV,dNdlnm_LMSV = obs_LMSV.mass_function()
plt.figure(figsize=(6,5))
plt.loglog(m,m*dNdlnm)
plt.loglog(m_MVJ,m_MVJ*dNdlnm_MVJ)
plt.loglog(m_LMSV,m_LMSV*dNdlnm_LMSV)
plt.xlim(1.e-7,3.e14)
plt.ylim(1.e7,1.e14)
plt.xlabel(r'$m$ [$M_{\odot}$]')
plt.ylabel(r'$m^{2}dN_{\rm sh}/dm$ [$M_{\odot}$]')
plt.legend(['Gaussian','MVJ, $f_{NL}=5000$','LMSV, $f_{NL}=5000$'])
plt.title(r'$M_{200}=10^{14}M_{\odot}$')
plt.show()
```
