```python
import numpy as np
import matplotlib.pyplot as plt
import sashimi_c


mdl = sashimi_c.halo_model()
M0 = 1e12*mdl.Msun
z = np.linspace(0,7)
mzi = mdl.Mzi(M0,z)
plt.plot(z,mzi)
plt.yscale('log')
plt.xlabel('Redshift z')
plt.ylabel(r'Halo mass $M(z)$ [$M_\odot$]')
plt.title(r'Halo mass accretion history $M(z)$ for $M_0=10^{12} M_\odot$')
plt.grid()
plt.show()
```

```python
mdl = sashimi_c.halo_model()
M0 = 1e12*mdl.Msun
for zi in np.linspace(0,6,4):
    z = np.linspace(zi,7)
    mzi = mdl.Mzi(M0,zi)  # 今M0なハローが、ziで持つ質量
    mzii = mdl.Mzzi(M0,z,zi) # ziでmziのハローがzで持つ質量
    plt.plot(z,mzii,label=f'zi={zi:.1f}')
    plt.scatter(zi,mzi,marker='x',label=f'M({zi:.1f})')
plt.legend()
plt.yscale('log')
plt.xlabel('Redshift z')
plt.ylabel(r'Halo mass $M(z)$ [$M_\odot$]')
plt.title(r'Halo mass accretion history $M(z)$ for $M_0=10^{12} M_\odot$')
plt.grid()
plt.show()
```
